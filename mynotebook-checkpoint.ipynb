```python
#sohil
#program number 7( A )
class Circle:
    #constructor:initializes the circle wiht a radius 
    def __init__(self,radius):
        self.radius = radius #store the radius value 

    #Method to calculate the area of the circle 
    def area(self):
        return 3.14*self.radius * self.radius #formula pie*r2

    #Method to calculate the perimeter of the circle 
    def perimeter(self):
        return 2*3.14 * self.radius #formula 2pie.r

#using the class 
#create an object of circle with radius 5
c1=Circle(5)
#print the area and the perimeter
print("Area of Circle :",c1.area())
print("Perimeter of Circle:", c1.perimeter())
```
